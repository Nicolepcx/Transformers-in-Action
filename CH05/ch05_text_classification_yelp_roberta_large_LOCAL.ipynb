# About this notebook

This Jupyter notebook performs a comprehensive analysis of text classification on the Yelp review dataset using both traditional machine learning and Transformers. Here are some key points covered in the notebook:

- __Baseline Model with Naive Bayes__: The notebook begins by creating a baseline model using Naive Bayes, a traditional machine learning algorithm frequently used for text classification tasks. This model serves as a benchmark for evaluating the performance of subsequent models.

- __TextClassifier for Transformer Models__: The notebook then utilizes the TextClassifier class to evaluate the performance of various Transformer models on the Yelp review  dataset "out-of-the-box" without any fine-tuning. This allows for an understanding of how well these pre-trained models perform on the task right away.

- __Fine-Tuning RoBERTa-large__: The notebook then fine-tunes the RoBERTa-large model. Fine-tuning involves training the pre-existing model on the Yelp review dataset, allowing the model to specialize its knowledge towards the text classification task at hand.

- __CustomTrainer for Balanced Training and Custom Learning Rate Scheduler__: Given the complex nature of the text classification task, the notebook uses a `CustomTrainer` class that helps handle class imbalance by using class weights, allowing the model to better learn the nuances in the different categories of the Yelp review dataset. Furthermore, the `CustomTrainer` class also supports the use of custom learning rate schedulers, specifically the `ReduceLROnPlateau` scheduler, to optimize the learning process.

- __Evaluation__: Lastly, we evaluate and compare the results from the fine-tuned RoBERTa model against the baseline Naive Bayes model and the initial out-of-the-box Transformer models' performance. This comparison is performed using a confusion matrix created from a classification report from the `sklearn` library, with a particular focus on the weighted F1-score and accuracy as key performance metrics. This detailed comparison provides insight into how much improvement the fine-tuning process brought about, and the relative strengths and weaknesses of each model for this particular task.

# Using NaÃ¯ve Bayes classifier as baseline

Naive Bayes methods are a set of supervised learning algorithms often used for text classification, spam detection, and other similar tasks. These algorithms are based on applying Bayes' theorem with the "naive" assumption of conditional independence between every pair of features given the value of the class variable.

Bayes' theorem states the following relationship, given class variable \(y\) and dependent feature vector \(x_1\) through \(x_n\):

$$
P(y | x_1, \ldots, x_n) = \frac{P(y) P(x_1, \ldots, x_n | y)}{P(x_1, \ldots, x_n)}
$$

Using the naive conditional independence assumption:

$$
P(x_i | y, x_1, \ldots, x_{i-1}, x_{i+1}, \ldots, x_n) = P(x_i | y),
$$

for all \(i\), this relationship is simplified to:

$$
P(y | x_1, \ldots, x_n) = \frac{P(y) \prod_{i=1}^n P(x_i | y)}{P(x_1, \ldots, x_n)}
$$

Now let's consider the case where the feature vector consists of words $(w_1, w_2, w_3, \dots, w_n)$. The formula for the Naive Bayes method becomes:

\begin{align*}
\underset{y \in\{\text {Pos,Neg}\}}{\arg \max } P\left(y \mid w_1, w_2, w_3, \dots, w_n\right) &=\underset{y \in\{\text {Pos,Neg}\}}{\arg \max } \frac{P\left(w_1, w_2, w_3, \dots, w_n \mid y\right) \cdot P(y)}{P\left(w_1, w_2, w_3, \dots, w_n\right)} \\
&=\underset{y \in\{\text {Pos,Neg}\}}{\arg \max } P\left(w_1, w_2, w_3, \dots, w_n \mid y\right) \cdot P(y) \\
&=\underset{y \in\{\text {Pos,Neg}\}}{\arg \max } P\left(w_1 \mid y\right) \cdot \prod_{i=2}^n P\left(w_i \mid w_1, \dots, w_{i-1}, y\right) \cdot P(y) \\
&=\underset{y \in\{\text {Pos,Neg}\}}{\arg \max } \prod_{i=1}^n P\left(w_i \mid y\right) \cdot P(y)
\end{align*}

In the second line, we remove the denominator $(P\left(w_1, w_2, w_3, \dots, w_n\right))$ as it's constant given the input. In the third line, we apply the chain rule of probability. Finally, in the fourth line, we use the naive conditional independence assumption to simplify the product of conditional probabilities.

## Custom trainer class

The CustomTrainer class extends the Hugging Face `Trainer` class, implementing additional functionality to suit specific needs for our machine learning task.

Here's a breakdown of its methods and their purpose:

- `__init__(*args, **kwargs)`: This constructor inherits from the Trainer's constructor. It also initializes a new attribute, best_metric, which will keep track of the best evaluation metric observed during the training process. This is useful for tasks where we want to save or perform an action based on the best model performance during training.

- `compute_loss(model, inputs, return_outputs=False)`: This method is used to compute the loss for a batch of inputs. Notably, it uses a weighted version of CrossEntropyLoss, which can be beneficial in cases where our dataset is imbalanced. This method takes into account the different class weights while computing the loss.

- `training_step(model, inputs)`: This method performs a single training step. It prepares the inputs, computes the loss, and performs the backward pass for gradient computation. It also handles different precision modes like fp16 or deepspeed if they are enabled in the training arguments.

- `training_epoch_end(outputs)`: This method is called at the end of each training epoch. It adds a custom function to evaluate the model and update the learning rate scheduler if the scheduler used is ReduceLROnPlateau. The learning rate is updated based on the evaluation metric. This is particularly useful when we want the learning rate to be adjusted based on the model's performance on a validation dataset, which can lead to better model performance and faster convergence.

By creating a `CustomTrainer`, we can effectively incorporate custom loss functions, training procedures, and end-of-epoch actions into the established Hugging Face training framework, giving us both the benefits of the robust `Trainer` class and the flexibility to meet our specific needs.

As for this case, when we want to use class weights. As a way to handle class imbalance in our dataset. This is especially useful when we have unevenly distributed classes or text with fine nuances to distinguish, and want to make sure your model does not overwhelmingly favor the majority class(es).

Class weights are assigned to the CrossEntropyLoss function in the compute_loss method of the CustomTrainer class. These weights provide a way for the model to pay more or less attention to specific classes during training.

In our example, we are assigning higher weights (1.5) to the 2nd, 3rd, and 4th classes, and standard weights (1.0) to the 1st and 5th classes. This means the model will pay more attention to getting the 2nd, 3rd, and 4th classes correct, potentially improving performance if these classes were previously being overshadowed by the other classes.

These weights are passed to the CustomTrainer through the class_weights tensor that we have created and then transferred to the appropriate device for computation (training_args.device).

This tensor is directly used in the CrossEntropyLoss function in the compute_loss method of the CustomTrainer class, enabling the model to consider these weights during the training process. This kind of class-specific weighting can greatly help in training more balanced and fair models, especially in scenarios where class imbalance is a significant issue.

```python
class CustomTrainer(Trainer):
    """
    CustomTrainer extends the Hugging Face Trainer class, adding custom functionality for training,
    loss computation and end of epoch actions.

    Attributes:
        best_metric (float): Stores the best evaluation metric observed during training.

    Methods:
        __init__(*args, **kwargs): Inherits Trainer's constructor and initializes best_metric.

        compute_loss(model: PreTrainedModel, inputs: Dict[str, Union[torch.Tensor, Any]],
        return_outputs=False) -> torch.Tensor:
            Computes the CrossEntropy loss using class weights.

        training_step(model: nn.Module, inputs: Dict[str, Union[torch.Tensor, Any]]) -> torch.Tensor:
            Runs a single training step, including loss computation and backward pass.

        training_epoch_end(outputs):
            At the end of each training epoch, evaluate the model if the learning rate scheduler is
            ReduceLROnPlateau and update the learning rate scheduler with the evaluation metric.
    """

    def __init__(self, *args, **kwargs):
        super().__init__(*args, **kwargs)
        self.best_metric = float('-inf')

    def compute_loss(self, model: PreTrainedModel, inputs: Dict[str, Union[torch.Tensor, Any]], return_outputs=False) -> torch.Tensor:
        labels = inputs.pop("labels")
        outputs = model(**inputs)
        logits = outputs.logits

        # Use the CrossEntropyLoss with class weights
        loss_fct = CrossEntropyLoss(weight=class_weights)
        loss = loss_fct(logits.view(-1, self.model.config.num_labels), labels.view(-1))

        return (loss, outputs) if return_outputs else loss

    def training_step(self, model: nn.Module, inputs: Dict[str, Union[torch.Tensor, Any]]) -> torch.Tensor:
        model.train()
        inputs = self._prepare_inputs(inputs)
        loss = self.compute_loss(model, inputs)
        loss = loss / self.args.gradient_accumulation_steps

        if self.args.fp16:
            self.scaler.scale(loss).backward()
        elif self.args.fp16_backend == "apex":
            with amp.scale_loss(loss, self.optimizer) as scaled_loss:
                scaled_loss.backward()
        elif self.args.deepspeed is not None:
            self.deepspeed.backward(loss)
        else:
            loss.backward()

        return loss.detach()

    def training_epoch_end(self, outputs):
        if self.lr_scheduler.__class__.__name__ == "ReduceLROnPlateau":
            eval_metrics = self.evaluate()
            metric_to_check = eval_metrics["eval_accuracy"]

            # Check if the metric improved
            if metric_to_check > self.best_metric:
                self.best_metric = metric_to_check

            # Update the learning rate scheduler with the evaluation metric
            self.lr_scheduler.step(metric_to_check)
        else:
            super().training_epoch_end(outputs)
```

```python

training_args = TrainingArguments(
    output_dir="./results",
    overwrite_output_dir=True,
    num_train_epochs=3,
    per_device_train_batch_size=batch_size,
    per_device_eval_batch_size=batch_size,
    save_strategy="epoch",
    save_total_limit=2,
    evaluation_strategy="epoch",
    learning_rate=5e-5,
    weight_decay=0.01,
    logging_strategy="epoch",
    logging_steps=10,
    load_best_model_at_end=True,
    warmup_steps=500,
)

class_weights = torch.tensor([1.0, 1.5, 1.5, 1.5, 1.0]).to(training_args.device)

trainer = CustomTrainer(
    model=model,
    args=training_args,
    train_dataset=train_dataset,
    eval_dataset=val_dataset,
    compute_metrics=compute_metrics,
    callbacks=[early_stopping_callback]
)


# Train the model
trainer.train()
```
